```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import ctrgcn_zhr
import torch
import math
import csv
import os
import shutil

txt_folder = '../../YOLO/datasets/train/labels/'

count = 0
with open('../../用于最终评估的各个矩阵和csv和图片/YOLO推理结果_遮挡一处.csv', 'r', newline='') as csv_file:
    reader = csv.reader(csv_file)
    output_iou = 0
    for row in reader:
        value_row = row[1:]
        
        # if count == 300:
        #     break

        if not os.path.exists(txt_folder+row[0][:-4]+'.txt'):
            continue
        count += 1
        txt_labels = []
        with open(txt_folder+row[0][:-4]+'.txt', 'r') as file:
            for line in file:
                line_labels = line.strip().split(' ')
                x3 = float(line_labels[1]) - float(line_labels[3])/2
                x4 = float(line_labels[1]) + float(line_labels[3])/2
                txt_labels.append([x3,x4])
        file.close()
        # print(txt_labels)

        final_iou = 0  # 每张图片对应的最终IOU结果
        if len(txt_labels) == 0:
            if len(value_row) == 0:
                final_iou = 1
            else:
                final_iou = 0
        else:
            final_intersection_length = 0
            final_union_length = 0
            for i in range(len(value_row)//5):
                conference = float(value_row[i*5])
                x1 = float(value_row[i*5+1]) - float(value_row[i*5+3])/2
                x2 = float(value_row[i*5+1]) + float(value_row[i*5+3])/2
                y1 = float(value_row[i*5+2]) - float(value_row[i*5+4])/2
                y2 = float(value_row[i*5+2]) + float(value_row[i*5+4])/2
                # print(conference,x1,x2,y1,y2)
                mid_intersection_length = 0
                mid_union_length = 0.000001     

                if conference > 0.38:
                    max_iou = 0
                    for j in range(len(txt_labels)):
                        intersection_start = max(x1, txt_labels[j][0])
                        intersection_end = min(x2, txt_labels[j][1])
                        intersection_length = max(0, intersection_end - intersection_start)
                        union_length = (x2 - x1) + (txt_labels[j][1] - txt_labels[j][0]) - intersection_length
                        iou = intersection_length / union_length
                        if max_iou<=iou:
                            max_iou = iou
                            mid_intersection_length = intersection_length
                            mid_union_length = union_length
                        max_iou = max(max_iou,iou)
                    # print(max_iou)

                    final_intersection_length += mid_intersection_length
                    final_union_length += mid_union_length
            if final_union_length == 0:
                final_iou = 1
            else:
                final_iou = final_intersection_length / final_union_length
        output_iou += final_iou
        print('final_iou: ',final_iou)
        # print('================================')
output_iou = output_iou / count
print("最终的iou结果: ",output_iou)
print(count)
```
